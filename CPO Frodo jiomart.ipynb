```python
def rider_data(riders):
    query = '''select
                ri.id as rider_id,
                
                case  
                    when ri.rider_type = 0  then 'Rider'
                    when ri.rider_type = 1  then 'Ops'
                    when ri.rider_type = 2  then 'Helper'
                    when ri.rider_type = 3  then 'Freelancer'
                    when ri.rider_type = 4  then 'RIDES'
                    when ri.rider_type = 5  then 'DC Staff'
                    when ri.rider_type = 6  then 'Wish'
                    when ri.rider_type = 7  then 'HULK'
                    when ri.rider_type = 8  then 'LCV'
                    when ri.rider_type = 9  then 'MILK_RUN'
                    when ri.rider_type = 10 then 'RIDER_X'
                    when ri.rider_type = 11 then 'SELLER'
                    when ri.rider_type = 12 then 'Vendor Ecom'
                    when ri.rider_type = 13 then 'PARENT_KIOSK'
                    when ri.rider_type = 14 then 'SELF_KIOSK'
                    when ri.rider_type = 15 then 'kiosk rider'
                    when ri.rider_type = 16 then 'Rider' else null end as rider_group,
              
              Case when rc.name in ('Delhi','Faridabad','Gurgaon','Ghaziabad','Noida') then 'NCR'
                    when rc.name = 'Bengaluru' then 'BLR'
                    when rc.name = 'Chennai' then 'CHN'
                    when rc.name = 'Pune' then 'PNQ'
                    when rc.name = 'Kolkata' then 'CCU'
                    when rc.name = 'Hyderabad' then 'HYD'
                    when rc.name in ('Mumbai','Navi Mumbai') then 'BOM'
                    else rc.name end as city,

                    case when rrp.vehicle_type = 3 then "cyclist" else "biker" end as vehicle_type
                    
                    
                from
                RMS_riderinfo ri
                left join RMS_riderpersonalinfo rrp on rrp.fk_rider_id = ri.id
                left join RMS_city rc on ri.city=rc.hl_city_id
            where ri.id in {riders} group by 1
            '''.format(riders = riders)
    df = sql.read_sql(query, cnx_rms)
    df['rider_id']=pd.to_numeric(df['rider_id'])
    return df

kiosk_riders = rider_data(rider_list)
kiosk_riders['rider_id'] = kiosk_riders['rider_id'].astype(str)
kiosk_riders
```

```python
df = rider_base.merge(kiosk_riders, left_on = ['rider_id'], right_on = ['rider_id'], how= 'left')
df
```

```python
def payout_structures(i_date,f_date):
    query = '''
    
SELECT

rpay.rider_id,
date_start,
date_end,
structure_name

from 
payout_engine_riderpayoutstructures as rpay 
left join payout_engine_payoutstructures as pay on rpay.payout_structure_id = pay.id

 
where
((date_end >= '{i_date}' and  date_start <= '{i_date}')
or (date_start <= '{f_date}' and date_end >= '{f_date}' ) or (date_start >= '{i_date}' and date_end <= '{f_date}' )) 

order by 2,1
                 
                 '''.format(i_date=i_date,f_date=f_date)
    df = sql.read_sql(query, cnx_smaug)
    return df



payout_structures = payout_structures(i_date,f_date)
payout_structures['rider_id'] = payout_structures['rider_id'].astype(str)
payout_structures
```

```python
def get_ecom_orders(row):
    try:
        value = json.loads(row)["total_postman_order_count"]
        return value
    except:
        return None


def get_ecom_order1(row):
    try:
        value = json.loads(row)["mkt_picked_count"]
        return value
    except:
        return None 


def get_ecom_order2(row):
    try:
        value = json.loads(row)["fwd_delivered_order_count"]
        return value
    except:
        return None


def get_ecom_order3(row):
    try:
        value = json.loads(row)["rev_picked_order_count"]
        return value
    except:
        return None


def get_ecom_order4(row):
    try:
        value = json.loads(row)["rts_delivered_order_count"]
        return value
    except:
        return None 

# Smaug_mysql = db.credentials['Smaug_mysql']
# cnx = sqlcon.connect(user=Smaug_mysql['user'], password=Smaug_mysql['password'], host=Smaug_mysql['host'],database=Smaug_mysql['database'])


cnx = sqlcon.connect(user='sf4036', password= 'LfAeNndsue', host='payout-rep1.prod.db.shadowfax.in',database='smaug_dataengine')

def get_rider_logins():
    query='''
    select rider_id,order_date as date,ecommerce_order_count as order_id,postman_order_count as smaug_postman_count
    from data_engine_riderleveldata ed where  order_date between '{i_date}' and '{f_date}'
    '''.format(i_date=i_date,f_date=f_date)
    df=pd.read_sql(query,cnx)
    df["smaug_postman_count"] = df["smaug_postman_count"].apply(get_ecom_orders)
    df["smaug_mkt_picked_count"] = df["order_id"].apply(get_ecom_order1)
    df["smaug_fwd_del_count"] = df["order_id"].apply(get_ecom_order2)
    df["smaug_rev_picked_count"] = df["order_id"].apply(get_ecom_order3)
    df["smaug_rts_del_count"] = df["order_id"].apply(get_ecom_order4)
    return df

cps_data=get_rider_logins()
cps_data = cps_data.drop_duplicates(subset=['rider_id','date'], keep='first')
cps_data = cps_data.fillna(0)
other_order = cps_data.copy()

other_order = other_order[['rider_id', 'date', 'smaug_postman_count',
       'smaug_mkt_picked_count', 'smaug_fwd_del_count',
       'smaug_rev_picked_count', 'smaug_rts_del_count']]

other_order['date']=other_order['date'].astype(str)
other_order['rider_id']=other_order['rider_id'].astype(str)
other_order
```

### Get All day's payout from Frodo DB ( component level)

```python
cnx_frodo = pg.connect( host = frodo['host'] , user = frodo['user'] ,password = frodo['password'] ,  database = frodo['database'])
i_date1 = i_date


def get_frodo_db(i_date,f_date):
    query = ''' select 
sfx_rider_id as rider_id,
payout_date as date,
rpc.payout_sub_title,
sum(case when rpc.payout_type  = 1 or rpc.payout_type  = 20  then amount end) as MG_PAY_frodo,
sum(case when rpc.payout_type  = 2  then amount end) as  OVERTIME_frodo,
sum(case when rpc.payout_type  = 3  then amount end) as  LWOA_FINE_frodo,
sum(case when rpc.payout_type  = 4  then amount end) as  INCENTIVES_frodo,
sum(case when rpc.payout_type  = 5  then amount end) as  ORDER_PAY_frodo,
sum(case when rpc.payout_type  = 6  then amount end) as  MBG_INCENTIVES_frodo,
sum(case when rpc.payout_type  = 7  then amount end) as  REFERRAL_BONUS_frodo,
sum(case when rpc.payout_type  = 8 and lower(payout_sub_title) not like '%reversal%'  then amount end) as  ADDITIONAL_EARNINGS_frodo,
sum(case when rpc.payout_type  = 9  then amount end) as  PENALTY_frodo,
sum(case when rpc.payout_type  = 10 then amount end) as  DEDUCTIONS_frodo,
sum(case when rpc.payout_type  = 11 then amount end) as  FULL_DISTANCE_PAY_frodo,
sum(case when rpc.payout_type  = 12 then amount end) as  FIRSTMILE_DISTANCE_PAY_frodo,
sum(case when rpc.payout_type  = 13 then amount end) as  LASTMILE_DISTANCE_PAY_frodo,
sum(case when rpc.payout_type  = 14 then amount end) as  WEIGHT_PAY_frodo,
sum(case when rpc.payout_type  = 15 then amount end) as  TRIP_DISTANCE_PAY_frodo,
sum(case when rpc.payout_type  = 16 then amount end) as  TRIP_WEIGHT_PAY_frodo,
sum(case when rpc.payout_type  = 17 then amount end) as  JOINING_BONUS_PAY_frodo, 
sum(case when rpc.payout_type  in (1,4,5,6,8,10,11,12,13,14,15,16,20) and lower(payout_sub_title) not like '%reversal%' then amount end) as  Net_day_earning,               
sum(case when rpc.payout_type  in (1,2,3,4,5,6,8,9,10,11,12,13,14,15,16,20) then amount end) as  Net_frodo,
sum(case when lower(rpc.payout_sub_title) like '%rain%' then amount end) as rain_pay
from payouts_riderpayoutcalculations rpc

where
rpc.payout_date between '{i_date}' and '{f_date}'

group by 1,2,3
order by 3

'''.format(i_date=i_date,f_date=f_date)
    df=pd.read_sql(query,cnx_frodo)
    return df



print(i_date1)
pin=pd.DataFrame()
frodo_db=pd.DataFrame()

#############################################################

for i in range(numdays+1):
    print(i)
    f_date1 = i_date1 + dt.timedelta(days=0)
    frodo = db.credentials['frodo']
    cnx_frodo = pg.connect( host = frodo['host'] , user = frodo['user'] ,password = frodo['password'] ,  database = frodo['database'])

    pin=get_frodo_db(i_date1,f_date1)   
    frodo_db=frodo_db.append(pin)
    
    print(i_date1)
    i_date1=f_date1+ dt.timedelta(days=1)
    print(f_date1)
############################################################

frodo_db = frodo_db.fillna(0)
frodo_db['date']=frodo_db['date'].astype(str)
frodo_db['rider_id']=frodo_db['rider_id'].astype(str)
frodo_db['payout_sub_title'] = frodo_db['payout_sub_title'].astype(str)
print("DONE")
frodo_db
```

```python
payout = frodo_db.loc[(frodo_db['payout_sub_title'].str.contains('jio', case = False))  ]
payout = pd.pivot_table(payout,index=["rider_id",'date'],values=['net_day_earning','order_pay_frodo','incentives_frodo','mg_pay_frodo','additional_earnings_frodo','full_distance_pay_frodo','firstmile_distance_pay_frodo','lastmile_distance_pay_frodo','trip_distance_pay_frodo','rain_pay','penalty_frodo','deductions_frodo'],aggfunc={'net_day_earning':sum,'order_pay_frodo':sum,'incentives_frodo':sum,'mg_pay_frodo':sum,'additional_earnings_frodo':sum,'full_distance_pay_frodo':sum,'firstmile_distance_pay_frodo':sum,'lastmile_distance_pay_frodo':sum,'trip_distance_pay_frodo':sum,'rain_pay':sum,'penalty_frodo':sum,'deductions_frodo':sum}).reset_index()
payout
```
